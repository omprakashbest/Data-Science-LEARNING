# • Plots under this section :-

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px  # for dataset Only
```

```python
# Dataset Load

tips = sns.load_dataset('tips') 
tips
```

```python
# Scatter Plot -> axis level function

sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='time', size='size')
```

```python
# Relplot -> figure level function

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', hue='sex', style='time', size='size')
```

```python
# Line Plot :- axis level function 

gap = px.data.gapminder()

temp_df = gap[gap['country'] == 'India'] # filtering the data for India only
temp_df
```

```python
sns.lineplot(x='year', y='lifeExp', data=temp_df)
```

```python
# figure level function 

sns.relplot(x='year', y='lifeExp', data=temp_df, kind='line')
```

```python
# hue -> style parameters

temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])] # checking gap dataset containing 
temp_df
```

```python
# figure level function:- 

sns.relplot(x='year', y='lifeExp', data=temp_df, kind='line', hue='country', style='continent', size='continent')
```

```python
# facet plot :- works with Only figure level functions 

# figure level 
sns.relplot(data=tips, x='total_bill', y='tip', kind='line', col='sex', row='smoker')
```

```python
# col wrap -> it will wrap the column into multiple rows according to the number of columns we have and the number of columns we want.

sns.relplot(x='lifeExp', y='gdpPercap',data=gap, kind='scatter', col='year', col_wrap=4)
```

# 2. Distribution Plots :-
    • Used for univariate analysis
    • Used to find out the distribution
    • Range of the observation
    • Central Tendency
    • Is the data bimodal?

# Plots under distribution plot :-
    • Histplot 
    • KDE Plot 
    • Rug Plot

```python
# figure level function -> distplot 
# axis level function ->  histplot, kdeplot, rugplot 
```

```python
# plotting univariate histogram 

# axis level function

sns.histplot(data=tips, x='total_bill')
```

```python
# figure level function 

sns.displot(data=tips, x='total_bill', kind='hist') 
```

```python
# bins parameter -> it will decide the number of bins we want in our histogram

sns.histplot(data=tips, x='total_bill', bins=20)
```

```python
# It's also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables.

# countplot -> it acting as a histogram for categorical variable but it not actual histogram.
sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue parameter

sns.displot(data=tips, x='tip', kind='hist', hue='sex')
```

```python
# element -> step -> parameter

sns.displot(data=tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
# dataset :
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x='age', kind='hist', hue='sex', element='step')
```

```python
# faceting using col and row parameters -> Only works with figure level functions

sns.displot(data=tips, x='tip', kind='hist', col='sex')
```

```python
# kde plot -> 
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel,
# producing a continuous density estimate. 
```

```python
# axis level function

sns.kdeplot(data=tips, x='total_bill')
```

```python
# figure level function

sns.displot(data=tips, x='total_bill', kind='kde')
```

```python
# hue -> fill -> parameters 

# height and aspect parameters -> it will change the size of the plot in figure level functions

sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True, height=5, aspect=2)
```

```python
# Rug Plot :- axis level function

# Plot marginal distribution by drawing ticks along the x and y axis.

# This function is intended to complement other plots by showing the location of individual observations in and unobtrusive way.


sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

```python
# Bivariate histogram 

# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count observations within each rectangle with the fill color.


sns.displot(data=tips, x='total_bill', y='tip', kind='hist') # figure level function
```

```python
# Bivariate KDE plot
# A bivariate KDE plot smooths the (x, y) observations with a 2D Gaussian function.

sns.displot(data=tips, x='total_bill', y='tip', kind='kde')
```

# 3. Matrix Plots :-
    • Heatmap
    • Cluster Map

```python
# Heatmap : axis level function  -> Only function .

# Plot rectangular data as a color-encoded matrix.


# converting the dataset into pivot table for wide format

temp_gap = gap.pivot(index='country', columns='year', values='lifeExp')


plt.figure(figsize=(15, 10)) # setting the size of the plot
sns.heatmap(temp_gap)
```

```python
# annot parameter -> it will show the value in each cell of heatmap
# linewidth parameter -> it will create the gap between the cells of heatmap
# cmap -> it will change the color of heatmap

temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country', columns='year', values='lifeExp')

plt.figure(figsize=(15, 15))
sns.heatmap(temp_df, annot=True, linewidths=0.5, cmap='summer')
```

```python
# axis level function -> clustermap

## clustermap -> it will create the cluster of the data and it also create the dendrogram for the data.

# Plot a matrix dataset as a hierarchically-clustered heatmap.
# this function requires scipy to be available.


# dataset :
iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:, [0, 1, 2, 3]]) # selecting the 4 columns of the dataset for clustering
```

# BoxPlot :
A box plot is a standardized way of displaying the distribution of data based on a five number summary(minimum",
"first quartile[Q1], median third quartile[Q3], "maximum"). It can tell you about your outliers adn what their values are. Boxplots can also tell you if you data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

```python
# BoxPlot : axis level function 

sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
# single boxplot -> numerical column

sns.boxplot(data=tips, y='total_bill')
```

# Violin Plot = (Box Plot + KDE Plot) :


```python
# Violin Plot : axis level function

sns.violinplot(data=tips, x='day', y='total_bill')
```

```python
# figure level function 

sns.catplot(data=tips, x='day', y='total_bill', kind='violin', height=4, aspect=2, hue='sex', split=True)
```

```python
# Categorical Estimate Plot -> Central tendency 
#     barplot
#     pointplot 
#     countplot
```

```python
# Bar plot -> axis level function

# When there are multiple observations in each category, it also uses bootstrapping to compute confidence interval
# around the estimate, give which is plotted using error bars.


# errorbar parameter -> it will remove the error bars from the plot

sns.barplot(data=tips, x='sex', y='total_bill', errorbar=None)
```

```python
# figure level function
import numpy as np 


sns.catplot(data=tips, x='sex', y='total_bill', kind='bar', estimator='mean')
```

```python
# Point Plot -> axis level function

# When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval
# around the estimate, give which is plotted using error bars.

sns.pointplot(data=tips, x='sex', y='total_bill', estimator='mean', hue='smoker')
```

```python
# Count Plot -> axis level function

# A special case for the bar plot when you want to show the number of observations in each category rather than
# computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative,
# variable.


sns.countplot(data=tips, x='sex', hue='day')
```

```python
# facet using col and row parameters -> Only works with figure level functions

sns.catplot(data=tips, x='sex', y='total_bill', kind='box', col='smoker', row='time')
```

# Regression Plots :
    • reg plot
    • lm plot

In the simplest Invocation, both functions draw a scatterplot of two variables, x and y, and then fit the 
regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
# axis level function 
# hue parameter not available in regplot but available in lmplot

sns.regplot(data=tips, x='total_bill', y='tip')
```

```python

sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
# residplot -> it will plot the residuals of the regression model

sns.residplot(data=tips, x='total_bill', y='tip')
```

# Multi-Plot Grids :-
    • Facet Grid 
    • Pair Plot 
    • Joint Plot

```python
# A second way to plot facet plots -> FacetGrid :-

# A FacetGrid can be used to draw any type of plot onto each facet, not just relational plots.

# one way :-
sns.catplot(data=tips, x='sex', y='total_bill', kind='violin', col='day', row='time')
```

```python
# Second way :- FacetGrid :- 

sns.FacetGrid(data=tips, col='day', row='time', hue='sex') \
.map(sns.scatterplot, 'tip', 'total_bill').add_legend()
```

# Plotting PairWise Relationships (PairGrid Vs PairPlot) :-


```python
# dataset : iris -> it is a multivariate dataset with 4 numerical columns and 1 categorical column.
# pairplot -> it will plot the pairwise relationships in a dataset.

sns.pairplot(data=iris, hue='species')
```

```python
# PairGrid -> it will create the pairwise relationships in a dataset but it also allows us to customize 
# the plots in the pairwise relationships.


sns.PairGrid(data=iris, hue='species') \
.map(sns.scatterplot) # function passing one at time
```

```python
# map_diag -> map_offdiag  # it will plot the diagonal and off-diagonal plots separately.

sns.PairGrid(data=iris, hue='species') \
.map_diag(sns.kdeplot, fill=True) \
.map_offdiag(sns.scatterplot)
```

```python
# map_diag -> map_upper -> map_lower 

sns.PairGrid(data=iris, hue='species') \
.map_diag(sns.kdeplot, fill=True) \
.map_upper(sns.scatterplot) \
.map_lower(sns.histplot)
```

```python
# vars -> it will select the columns of the dataset for plotting the pairwise relationships.

sns.PairGrid(data=iris, hue='species', vars=['sepal_width', 'petal_width']) \
.map_diag(sns.kdeplot, fill=True) \
.map_upper(sns.scatterplot) \
.map_lower(sns.histplot)
```

# JointGrid Vs JointPlot :-
    • JointGrid -> it will create the joint relationships in a dataset but it also allows us to customize 
    the plots in the joint relationships.

```python
# JointPlot :-
# kind parameter -> it will change the type of the plot in the joint relationships.(by default it is scatter plot)

sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex')
```

```python
# JointGrid :-

sns.JointGrid(data=tips, x='total_bill', y='tip', hue='sex') \
.plot(sns.scatterplot, sns.kdeplot) # passing the functions for joint relationships
```
