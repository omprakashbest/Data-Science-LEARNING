# 1. Relational Plots :-
    • To see the statistical relation between two or more variables
    • Bivariate Plots 

# • Plots under this section :- 
    • Scatter Plot 
    • Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px  # for dataset Only
```

```python
# Dataset Load

tips = sns.load_dataset('tips')
tips
```

```python
# Scatter Plot -> axis level function

sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='time', size='size')
```

```python
# Relplot -> figure level function

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', hue='sex', style='time', size='size')
```

```python
# Line Plot :- axis level function 

gap = px.data.gapminder()

temp_df = gap[gap['country'] == 'India'] # filtering the data for India only
temp_df
```

```python
sns.lineplot(x='year', y='lifeExp', data=temp_df)
```

```python
# figure level function 

sns.relplot(x='year', y='lifeExp', data=temp_df, kind='line')
```

```python
# hue -> style parameters

temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])] # checking gap dataset containing 
temp_df
```

```python
# figure level function:- 

sns.relplot(x='year', y='lifeExp', data=temp_df, kind='line', hue='country', style='continent', size='continent')
```

```python
# facet plot :- works with Only figure level functions 

# figure level 
sns.relplot(data=tips, x='total_bill', y='tip', kind='line', col='sex', row='day')
```

```python
# col wrap -> it will wrap the column into multiple rows according to the number of columns we have and the number of columns we want.

sns.relplot(x='lifeExp', y='gdpPercap',data=gap, kind='scatter', col='year', col_wrap=4)
```

# 2. Distribution Plots :-
    • Used for univariate analysis
    • Used to find out the distribution
    • Range of the observation
    • Central Tendency
    • Is the data bimodal?
    • Are there outliers?

# Plots under distribution plot :-
    • Histplot 
    • KDE Plot 
    • Rug Plot

```python
# figure level function -> distplot 
# axis level function ->  histplot, kdeplot, rugplot 
```

```python
# plotting univariate histogram 

# axis level function

sns.histplot(data=tips, x='total_bill')
```

```python
# figure level function 

sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
# bins parameter -> it will decide the number of bins we want in our histogram

sns.histplot(data=tips, x='total_bill', bins=20)
```

```python
# It's also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables.

# countplot -> it acting as a histogram for categorical variable but it not actual histogram.
sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue parameter

sns.displot(data=tips, x='tip', kind='hist', hue='sex')
```

```python
# element -> step -> parameter

sns.displot(data=tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
# dataset :
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x='age', kind='hist', hue='sex', element='step')
```

```python
# faceting using col and row parameters -> Only works with figure level functions

sns.displot(data=tips, x='tip', kind='hist', col='sex')
```

```python
# kde plot -> 
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel,
# producing a continuous density estimate. 
```

```python

sns.kdeplot(data=tips, x='total_bill')
```
