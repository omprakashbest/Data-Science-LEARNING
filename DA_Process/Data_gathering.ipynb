# 1. Data Gathering :-

### Importing from CSV :- 

```python
import pandas as pd
```

```python
# Opening with local csv file.

imdb = pd.read_csv('D:\LEARNING\imdb.csv')
imdb.head()
```

```python
# Opening a csv file from an URL(uniform Resource Locator).

# The library used to fetch the data from the URL is requests, 
# which allows us to send HTTP requests and handle responses.
import requests
from io import StringIO as str 

# The URL of the CSV file to be fetched.
url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
response = requests.get(url, headers=headers)
data = str(response.text) # StringIO is used to read the response text as a file-like object.

pd.read_csv(data)
```

```python
# Sep Parameter :- for tsv file (tab separated values).
# name parameter -> to give a name to the column of the data frame.

pd.read_csv('D:\LEARNING\sep.tsv', sep='\t', names=['ID', 'Name', 'Age', 'City'])
```

```python
# Index_col parameter -> to set a specific column as the index of the DataFrame.

pd.read_csv('D:\LEARNING\imdb.csv', index_col='Series_Title')
```

```python
# Header parameter -> to specify the row number to use as the column names, 
# or to indicate that there is no header row.

pd.read_csv('D:\LEARNING\imdb.csv', header=1)
```

```python
# usecols parameter -> to specify a subset of columns(Jo Jo cols chahiye mujhe) to read from the CSV file.

pd.read_csv('D:\LEARNING\imdb.csv', usecols=['Series_Title', 'IMDB_Rating'])
```

```python
# squeeze parameter but now it become function -> to reduce the dimensionality of the returned object when possible.
# also for convert the data frame into a series if the data frame has only one column.

pd.read_csv('D:\LEARNING\imdb.csv', usecols=['Series_Title']).squeeze()
```

```python
# skiprows/nrows parameter :
# nrows -> to specify the number of rows to read from the CSV file.

pd.read_csv('D:\LEARNING\practice_df.csv', nrows=50)
```

```python
# Encoding parameter -> to specify the encoding of the CSV file, 
# which is important for correctly reading files with special characters or non-ASCII text.

pd.read_csv('D:\LEARNING\imdb.csv', encoding='latin-1')
```

```python
# skip bad lines :

pd.read_csv('D:\\LEARNING\\imdb.csv', on_bad_lines='skip') # on_bad_lines='skip' in newer versions of pandas
```

```python
# dtype parameter :

pd.read_csv('D:\\LEARNING\\imdb.csv', dtype={'Runtime': float}).info()
```

```python
# Handling Dates :-
# parse_dates parameter

pd.read_csv('D:\LEARNING\ipl-matches.csv', parse_dates=['Date']).info()
```

```python
# convertors :-

# function :
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
    
pd.read_csv('D:\LEARNING\ipl-matches.csv', converters={'Team1': rename})
```

```python
# na_values parameter : jis jis word jo NaN consider krne chahte ho.

pd.read_csv("D:\LEARNING\imdb.csv", na_values=['Drama'])
```

```python
# Loading a huge dataset in chunks :

dfs = pd.read_csv("D:\LEARNING\movies.csv", chunksize=400)

# for print the dataset chunks you need to use For loop

for chunk in dfs:
    print(chunk.shape)
```
